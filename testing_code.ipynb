```python
import pandas as pd
import numpy as np
import os

# the following code is to test unity_data_extraction, didn't know where else to put it
kitchen_UO = pd.read_csv('unity_csvs/kitchen/mem/kitchenmem1/UnityObjects.csv')
kitchen_UO_agg = kitchen_UO.groupby('Type').agg({
    'Allocated(MB)': ['sum', 'count', 'mean'],
    'Resident(MB)': ['sum', 'mean']
}).reset_index()
kitchen_UO_agg.columns = [
        'Type',
        'allocated_sum', 'allocated_count', 'allocated_mean',
        'resident_sum', 'resident_mean'
    ]
kitchen_UO_agg
```

```python
kitchen_UO_agg2 = kitchen_UO.groupby('NameOfObject').agg({
    'Allocated(MB)': ['sum', 'count', 'mean'],
    'Resident(MB)': ['sum', 'mean']
}).reset_index()
kitchen_UO_agg2.columns = [
        'Type',
        'allocated_sum', 'allocated_count', 'allocated_mean',
        'resident_sum', 'resident_mean'
    ]
kitchen_UO_agg2
```
