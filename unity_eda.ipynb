```python
import pandas as pd
import numpy as np
import os
```

```python
kitchen_graphics = pd.read_csv('unity_csvs/kitchen/mem/kitchenmem1/Graphics.csv')
kitchen_graphics
```

```python
hallway_graphics = pd.read_csv('unity_csvs/hallway/mem/hallmem1/Graphics.csv')
hallway_graphics
```

```python
kitchen_only = kitchen_graphics[~kitchen_graphics['NameOfObject'].isin(hallway_graphics['NameOfObject'])]
print(f"Count: {len(kitchen_only)}")
print(kitchen_only)
```

```python
hallway_only = hallway_graphics[~hallway_graphics['NameOfObject'].isin(kitchen_graphics['NameOfObject'])]
print(f"Count: {len(hallway_only)}")
print(hallway_only)
```

```python
common_names = set(kitchen_graphics['NameOfObject']) & set(hallway_graphics['NameOfObject'])
print(f"Count: {len(common_names)}")
print(common_names)
```

```python
graphics_comparison = kitchen_graphics.merge(
    hallway_graphics, 
    on='NameOfObject', 
    how='inner', 
    suffixes=('_kitchen', '_hallway')
)
graphics_comparison
```

```python
memory_differs = graphics_comparison[
    graphics_comparison['Allocated(MB)_kitchen'] != graphics_comparison['Allocated(MB)_hallway']
]
memory_differs
```

```python
kitchen_graphics2 = pd.read_csv('unity_csvs/kitchen/mem/kitchenmem2/Graphics.csv')
kitchen_comparison = kitchen_graphics.merge(
    kitchen_graphics2, 
    on='NameOfObject', 
    how='inner', 
    suffixes=('_kitchen1', '_kitchen2')
)
memory_differs_kitch = kitchen_comparison[
    kitchen_comparison['Allocated(MB)_kitchen1'] != kitchen_comparison['Allocated(MB)_kitchen2']
]
memory_differs_kitch
```

```python
kitchen_agg = kitchen_graphics.groupby('NameOfObject').agg({
    'Allocated(MB)': ['sum', 'count', 'mean']
}).reset_index()
kitchen_agg.columns = ['NameOfObject', 'total_allocated', 'count', 'mean_allocated']
kitchen_agg
```

```python
hallway_agg = hallway_graphics.groupby('NameOfObject').agg({
    'Allocated(MB)': ['sum', 'count', 'mean']
}).reset_index()
hallway_agg.columns = ['NameOfObject', 'total_allocated', 'count', 'mean_allocated']
hallway_agg
```

```python
agg_comparison = kitchen_agg.merge(
    hallway_agg, 
    on='NameOfObject', 
    how='outer', 
    suffixes=('_kitchen', '_hallway')
)
agg_comparison = agg_comparison.fillna(0)
```

```python
memory_differs_agg = agg_comparison[
    (agg_comparison['total_allocated_kitchen'] != agg_comparison['total_allocated_hallway']) |
    (agg_comparison['count_kitchen'] != agg_comparison['count_hallway']) |
    (agg_comparison['mean_allocated_kitchen'] != agg_comparison['mean_allocated_hallway'])
]
memory_differs_agg
```
