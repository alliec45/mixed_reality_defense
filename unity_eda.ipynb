```python
# reusable function to streamline this process
import pandas as pd
import numpy as np

# dictionary mapping room names to their file paths
room_paths = {
    'kitchen': 'unity_csvs/kitchen/mem/kitchenmem',
    'hallway': 'unity_csvs/hallway/mem/hallmem',
    'lab': 'unity_csvs/lab/mem/labmem',
    'blinds_up': 'unity_csvs/blinds_up/mem/bumem',
    'blinds_down': 'unity_csvs/blinds_down/mem/bdmem'
}

def load_and_aggregate_AMO(filepath):
    df = pd.read_csv(filepath)
    agg = df.groupby('Type').agg({
        'Allocated(MB)': ['sum', 'count', 'mean'],
        'Count': ['sum', 'mean', 'min', 'max'],
        'Resident(MB)': ['sum', 'mean']
    }).reset_index()
    agg.columns = [
        'Type', 
        'allocated_sum', 'allocated_count', 'allocated_mean',
        'count_sum', 'count_mean', 'count_min', 'count_max',
        'resident_sum', 'resident_mean'
    ]
    return agg

def compare_two_rooms(agg1, agg2, name1, name2, key_col='Type'):
    comparison = agg1.merge(
        agg2,
        on=key_col,
        how='outer',
        suffixes=(f'_{name1}', f'_{name2}')
    ).fillna(0)
    
    # find differences
    cols1 = sorted([col for col in comparison.columns if col.endswith(f'_{name1}')])
    cols2 = sorted([col for col in comparison.columns if col.endswith(f'_{name2}')])
    differences = (comparison[cols1].values != comparison[cols2].values).any(axis=1)
    
    return comparison, comparison[differences]

def compare_rooms_AMO(room1, room2, scan_num=1):
    # build file paths
    path1 = f"{room_paths[room1]}{scan_num}/AllManagedObjects.csv"
    path2 = f"{room_paths[room2]}{scan_num}/AllManagedObjects.csv"
    
    # load and aggregate
    agg1 = load_and_aggregate_AMO(path1)
    agg2 = load_and_aggregate_AMO(path2)
    
    # compare
    full_comparison, differences = compare_two_rooms(agg1, agg2, room1, room2)
    
    print(f"{room1.upper()} vs {room2.upper()} (scan {scan_num})")
    print(f"Total types compared: {len(full_comparison)}")
    print(f"Types with differences: {len(differences)}")
    
    return full_comparison, differences

full_comp, diff = compare_rooms_AMO('kitchen', 'hallway', scan_num=1)
diff
```

```python
from collections import Counter

# compare all room combinations and track differing Types
rooms = ['kitchen', 'hallway', 'lab', 'blinds_up', 'blinds_down']

type_diff_counter = Counter()
type_diff_details = {}

for j in range(1, 6):
    for i, room1 in enumerate(rooms):
        for room2 in rooms[i+1:]:
            full_comp, diff = compare_rooms_AMO(room1, room2, scan_num=j)
            
            differing_types = diff['Type'].tolist()
            
            type_diff_counter.update(differing_types)
            
            for t in differing_types:
                if t not in type_diff_details:
                    type_diff_details[t] = []
                type_diff_details[t].append((room1, room2, j))

type_summary = pd.DataFrame([
    {
        'Type': t,
        'diff_count': count,
        'room_pairs': type_diff_details[t]
    }
    for t, count in type_diff_counter.items()
]).sort_values('diff_count', ascending=False)

pairs_per_trial = len(rooms) * (len(rooms) - 1) // 2 # 10
max_comparisons = pairs_per_trial * 5 # 50

print(f"\nTotal unique Types that differ: {len(type_summary)}")
type_summary
```

```python
type_summary['diff_percentage'] = (type_summary['diff_count'] / max_comparisons * 100).round(2)
threshold = max_comparisons // 2 # 25
frequently_differ = type_summary[type_summary['diff_count'] >= threshold]
print(f"Types that differ in at least {threshold} comparisons (at least 50% of comparisons)")
print(f"Count: {len(frequently_differ)}")
frequently_differ
```

```python
def compare_cpu_files(filepath1, filepath2, name1, name2, pct_threshold=10):
    df1 = pd.read_csv(filepath1, sep=';')
    df2 = pd.read_csv(filepath2, sep=';')
    
    df1.columns = df1.columns.str.strip()
    df2.columns = df2.columns.str.strip()
    
    cpu_comparison = df1.merge(
        df2,
        on='Name',
        how='inner',
        suffixes=(f'_{name1}', f'_{name2}')
    ).fillna(0)
    
    cpu_comparison.columns = cpu_comparison.columns.str.strip()
    
    col1 = f'Median Time_{name1}'
    col2 = f'Median Time_{name2}'
    
    max_val = np.maximum(
        cpu_comparison[col1].abs(),
        cpu_comparison[col2].abs()
    )
    
    pct_diff = np.where(
        max_val > 0,
        abs(cpu_comparison[col1] - cpu_comparison[col2]) / max_val * 100,
        0
    )
    
    cpu_comparison['pct_diff'] = pct_diff
    
    flagged_df = cpu_comparison[cpu_comparison['pct_diff'] > pct_threshold][
        ['Name', col1, col2, 'pct_diff']
    ].sort_values('pct_diff', ascending=False)
    
    print(f"{name1.upper()} vs {name2.upper()}")
    print(f"Total rows compared: {len(cpu_comparison)}")
    print(f"Total flags (>{pct_threshold}% diff): {len(flagged_df)}")
    
    return cpu_comparison, flagged_df
```

```python
def compare_cpu_median(filepath1, filepath2, name1, name2, pct_threshold=10):
    df1 = pd.read_csv(filepath1, sep=';')
    df2 = pd.read_csv(filepath2, sep=';')
    
    df1.columns = df1.columns.str.strip()
    df2.columns = df2.columns.str.strip()
    
    cpu_comparison = df1.merge(
        df2,
        on='Name',
        how='inner',
        suffixes=(f'_{name1}', f'_{name2}')
    ).fillna(0)
    
    cpu_comparison.columns = cpu_comparison.columns.str.strip()
    
    col1 = f'Median Time_{name1}'
    col2 = f'Median Time_{name2}'
    
    max_val = np.maximum(cpu_comparison[col1].abs(), cpu_comparison[col2].abs())
    
    pct_diff = np.where(
        max_val > 0,
        abs(cpu_comparison[col1] - cpu_comparison[col2]) / max_val * 100,
        0
    )
    
    cpu_comparison['pct_diff'] = pct_diff
    
    flagged = cpu_comparison[cpu_comparison['pct_diff'] > pct_threshold][['Name', 'pct_diff']]
    
    return set(flagged['Name'].tolist())

def evaluate_threshold(threshold, num_trials=5):
    different_room_flags = Counter()
    same_room_flags = Counter()
    
    different_room_comparisons = 0
    same_room_comparisons = 0
    
    group_names = list(room_groups.keys())
    
    for trial in range(1, num_trials + 1):
        for i, group1 in enumerate(group_names):
            for group2 in group_names[i+1:]:
                for room1 in room_groups[group1]:
                    for room2 in room_groups[group2]:
                        try:
                            path1 = f"{cpu_paths[room1]}{trial}.csv"
                            path2 = f"{cpu_paths[room2]}{trial}.csv"
                            flagged = compare_cpu_median(path1, path2, room1, room2, threshold)
                            different_room_flags.update(flagged)
                            different_room_comparisons += 1
                        except Exception as e:
                            pass
        
        for group in group_names:
            rooms_in_group = room_groups[group]
            for i, room1 in enumerate(rooms_in_group):
                for room2 in rooms_in_group[i+1:]:
                    try:
                        path1 = f"{cpu_paths[room1]}{trial}.csv"
                        path2 = f"{cpu_paths[room2]}{trial}.csv"
                        flagged = compare_cpu_median(path1, path2, room1, room2, threshold)
                        same_room_flags.update(flagged)
                        same_room_comparisons += 1
                    except Exception as e:
                        pass
    
    return different_room_flags, same_room_flags, different_room_comparisons, same_room_comparisons
```

```python
# test one threshold value
threshold = 10
diff_flags, same_flags, diff_count, same_count = evaluate_threshold(threshold, num_trials=5)

print(f"Threshold: {threshold}%")
print(f"Different room comparisons: {diff_count}")
print(f"Same room comparisons: {same_count}")

all_names = set(diff_flags.keys()) | set(same_flags.keys())

results = []
for name in all_names:
    diff_rate = diff_flags[name] / diff_count * 100 if diff_count > 0 else 0
    same_rate = same_flags[name] / same_count * 100 if same_count > 0 else 0
    
    results.append({
        'Name': name,
        'diff_room_flags': diff_flags[name],
        'same_room_flags': same_flags[name],
        'diff_room_rate': round(diff_rate, 2),
        'same_room_rate': round(same_rate, 2),
        'discrimination_score': round(diff_rate - same_rate, 2)  # Higher = better feature
    })

results_df = pd.DataFrame(results).sort_values('discrimination_score', ascending=False)

print("\nBEST FEATURES (high diff_room_rate, low same_room_rate)")
results_df.head(30)

print("\nWORST FEATURES (flagged often in same room comparisons)")
results_df.tail(20)
```

```python
thresholds_to_test = [5, 10, 15, 20, 25, 30]

for threshold in thresholds_to_test:
    diff_flags, same_flags, diff_count, same_count = evaluate_threshold(threshold, num_trials=3)
    
    good_features = 0
    for name in set(diff_flags.keys()) | set(same_flags.keys()):
        diff_rate = diff_flags[name] / diff_count * 100 if diff_count > 0 else 0
        same_rate = same_flags[name] / same_count * 100 if same_count > 0 else 0
        if diff_rate > 50 and same_rate < 20: # good discrimination criteria
            good_features += 1
    
    print(f"Threshold {threshold}%: {good_features} good features")
```

```python
# Load original data
df = pd.read_csv('room_classification_features.csv')

# Separate blinds_up and blinds_down samples
blinds_up_df = df[df['room_label'] == 'blinds_up'].drop(columns=['room', 'room_label', 'trial'])
blinds_down_df = df[df['room_label'] == 'blinds_down'].drop(columns=['room', 'room_label', 'trial'])

print(f"Blinds up samples: {len(blinds_up_df)}")
print(f"Blinds down samples: {len(blinds_down_df)}")

# Calculate mean values for each feature
blinds_up_mean = blinds_up_df.mean()
blinds_down_mean = blinds_down_df.mean()

# Calculate percentage difference
comparison = pd.DataFrame({
    'blinds_up_mean': blinds_up_mean,
    'blinds_down_mean': blinds_down_mean,
})

# Avoid division by zero
max_val = np.maximum(comparison['blinds_up_mean'].abs(), comparison['blinds_down_mean'].abs())
comparison['pct_diff'] = np.where(
    max_val > 0,
    abs(comparison['blinds_up_mean'] - comparison['blinds_down_mean']) / max_val * 100,
    0
)

comparison = comparison.sort_values('pct_diff', ascending=False)

print("\n=== TOP 30 FEATURES THAT DIFFER BETWEEN BLINDS_UP AND BLINDS_DOWN ===")
print(comparison.head(30))

# Also check standard deviation to see consistency
blinds_up_std = blinds_up_df.std()
blinds_down_std = blinds_down_df.std()

comparison['blinds_up_std'] = blinds_up_std
comparison['blinds_down_std'] = blinds_down_std

# Features with high difference AND low variance (consistent difference)
comparison['discrimination_score'] = comparison['pct_diff'] / (comparison['blinds_up_std'] + comparison['blinds_down_std'] + 0.001)

print("\n=== BEST DISCRIMINATING FEATURES (high diff, low variance) ===")
print(comparison.sort_values('discrimination_score', ascending=False).head(20))
```

```python
# Load all blinds_up and blinds_down CPU files
import glob
import os

# Collect Median Time for key features across all trials
features_to_check = [
    'ConstraintManager::Update::JobSetup',
    'PostLateUpdate.BatchModeUpdate',
    'AddDirectionalLights',
    'Inl_UpdateVolumeFramework',
    'EarlyUpdate.GpuTimestamp',
    'Inl_Setup Forward Lights',
    'Inl_Setup Light Constants',
]

blinds_up_data = {f: [] for f in features_to_check}
blinds_down_data = {f: [] for f in features_to_check}

# Load all blinds_up trials
trial = 1
while True:
    path = f"unity_csvs/blinds_up/cpu/bu{trial}.csv"
    if os.path.exists(path):
        df = pd.read_csv(path, sep=';')
        df.columns = df.columns.str.strip()
        for feature in features_to_check:
            val = df[df['Name'] == feature]['Median Time'].values
            if len(val) > 0:
                blinds_up_data[feature].append(val[0])
        trial += 1
    else:
        break

# Also check motion/person/object variants
for variant in ['blinds_up_motion/cpu/bum', 'blinds_up_person/cpu/bup', 'blinds_up_object/cpu/buo']:
    trial = 1
    while True:
        path = f"unity_csvs/{variant}{trial}.csv"
        if os.path.exists(path):
            df = pd.read_csv(path, sep=';')
            df.columns = df.columns.str.strip()
            for feature in features_to_check:
                val = df[df['Name'] == feature]['Median Time'].values
                if len(val) > 0:
                    blinds_up_data[feature].append(val[0])
            trial += 1
        else:
            break

# Load all blinds_down trials
trial = 1
while True:
    path = f"unity_csvs/blinds_down/cpu/bd{trial}.csv"
    if os.path.exists(path):
        df = pd.read_csv(path, sep=';')
        df.columns = df.columns.str.strip()
        for feature in features_to_check:
            val = df[df['Name'] == feature]['Median Time'].values
            if len(val) > 0:
                blinds_down_data[feature].append(val[0])
        trial += 1
    else:
        break

# Also check motion/person/object variants
for variant in ['blinds_down_motion/cpu/bdm', 'blinds_down_person/cpu/bdp', 'blinds_down_object/cpu/bdo']:
    trial = 1
    while True:
        path = f"unity_csvs/{variant}{trial}.csv"
        if os.path.exists(path):
            df = pd.read_csv(path, sep=';')
            df.columns = df.columns.str.strip()
            for feature in features_to_check:
                val = df[df['Name'] == feature]['Median Time'].values
                if len(val) > 0:
                    blinds_down_data[feature].append(val[0])
            trial += 1
        else:
            break

# Compare distributions
print("=== FEATURE DISTRIBUTIONS: BLINDS_UP vs BLINDS_DOWN ===\n")
for feature in features_to_check:
    bu_vals = blinds_up_data[feature]
    bd_vals = blinds_down_data[feature]
    
    bu_mean = np.mean(bu_vals) if bu_vals else 0
    bu_std = np.std(bu_vals) if bu_vals else 0
    bu_min = np.min(bu_vals) if bu_vals else 0
    bu_max = np.max(bu_vals) if bu_vals else 0
    
    bd_mean = np.mean(bd_vals) if bd_vals else 0
    bd_std = np.std(bd_vals) if bd_vals else 0
    bd_min = np.min(bd_vals) if bd_vals else 0
    bd_max = np.max(bd_vals) if bd_vals else 0
    
    # Check if distributions overlap
    overlap = (bu_min <= bd_max) and (bd_min <= bu_max)
    
    print(f"--- {feature} ---")
    print(f"  Blinds Up:   mean={bu_mean:.6f}, std={bu_std:.6f}, range=[{bu_min:.6f}, {bu_max:.6f}]")
    print(f"  Blinds Down: mean={bd_mean:.6f}, std={bd_std:.6f}, range=[{bd_min:.6f}, {bd_max:.6f}]")
    print(f"  Distributions overlap: {overlap}")
    print()
```
